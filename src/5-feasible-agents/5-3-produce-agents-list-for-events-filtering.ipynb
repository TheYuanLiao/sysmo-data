# Agent list for further filtering
In this notebook, all agents' status, feasible/infeasible, are combined for further processing events and plans.

```python
import pandas as pd
df1 = pd.read_csv('combined_Palt.csv')
df2 = pd.read_csv('combined_Västra_Götaland.csv')
df3 = pd.read_csv('plans_samm_feasible_agents.csv')
df4 = pd.read_csv('plans_skane_feasible_agents.csv')
df5 = pd.read_csv('plans_sydost_feasible_agents.csv')
```

```python
combined_df = pd.concat([df1, df2, df3, df4, df5], ignore_index=True)
```

```python
combined_df.to_csv('combined_Sweden.csv', index=False)
```

```python
df = pd.read_csv('combined_Sweden.csv')
df    #Total no. of Feasible Agents for the Entire Sweden
```

```python
df['agent_status'] = 'feasible'
```
